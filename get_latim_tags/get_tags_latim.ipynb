## Separando cada frase/sentenca do texto

```
def split_sentences(latim_text):
    '''
    Separa as sentencas de acordo com as tags que marcam os fins delas. Retorna
    uma lista com cada uma das sentencas do texto como elementos
    '''
    sentences = []

    while latim_text is not "":
        sentences.append((latim_text[:latim_text.find("SENT") + len("SENT")])\
            .strip())
        latim_text = latim_text[latim_text.find("SENT") + len("SENT"):]

    return sentences


sentences = split_sentences(latim_text)

print(sentences)
print("Numero de sentencas encontradas: ", len(sentences))
```

## Definindo as tags que serao procuradas
Aqui, voce deve alterar o conteudo das variaveis tag_1, tag_2, ..., etc para as tags que voce quer procurar no texto. Deixe vazia as variavies que voce nao for utilizar

```
tag_1 = "N:dat"
tag_2 = "V:IND"
tag_3 = "N:nom"
tag_4 = ""
tag_5 = ""
tag_6 = ""


tag_not_wanted_1 = "N:acc"
tag_not_wanted_2 = ""
tag_not_wanted_3 = ""
tag_not_wanted_4 = ""
tag_not_wanted_5 = ""
tag_not_wanted_6 = ""


# se precisar da mais tags, crie uma outra variavel e acrescente ela nessa lista
tags = [tag_1, tag_2, tag_3, tag_4, tag_5, tag_6]
tags_not_wanted = [tag_not_wanted_1, tag_not_wanted_2, tag_not_wanted_3, \
                   tag_not_wanted_4, tag_not_wanted_5, tag_not_wanted_5]

# filtrando as tags vazias
tags = list(filter(None, tags))
tags_not_wanted = list(filter(None, tags_not_wanted))

print("tags que serao procuradas no texto: ", tags)
print("tags que serao excluidas, caso aparecam :", tags_not_wanted)
```

Execute isso para encontrar todas as tags que voce **quer**

```
def search_tag(sentences, tag):
  matching = [sent for sent in sentences if str(tag) in sent]

  return matching


matching = sentences

for tag in tags:
  matching = search_tag(matching, tag)

if matching == []:
  print("Nada encontrado") 

else:
  print("matching: ", matching)
  print("numero de sentencas encontradas: ", len(matching))
```

```
for sentence in matching:
  print(sentence, "\n")
```

Caso nao queira alguma tag, **execute isso depois**

```
def dont_want(matching, tag_not_wanted):
  new_matching = [sent for sent in matching if str(tag) not in sent]

  return new_matching

new_matching = matching

for tag_not_wanted in tags_not_wanted:
  new_matching = dont_want(new_matching, tags_not_wanted)

if new_matching == []:
  print("Nada encontrado!") 

else:
  print(new_matching)
  print("novo numero de tags encontradas, com nova condicao: ", len(new_matching))
```

```
for sentence in new_matching:
  print(sentence, "\n")
```
